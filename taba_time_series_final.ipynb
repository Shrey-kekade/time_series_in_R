# ARIMA

## Test of Stationarity : ADF Test

even though auto.arima manages stationarity, it's still a good practice to perform an initial stationarity check on the time series.

The Augmented Dickey-Fuller (ADF) test is commonly used to test for stationarity

```r
# Perform the Augmented Dickey-Fuller test
adf_test_result <- adf.test(train, alternative = "stationary")

# Display the result
print(adf_test_result)
```

The ADF test's null hypothesis states that the time series is non-stationary. The alternative hypothesis is that the series is stationary.

- **P-value**: The p-value of 0.1256 is greater than the commonly used threshold of 0.05. This means do not have enough evidence to reject the null hypothesis, indicating that the series is non-stationary.

## Auto ARIMA Method

The ARIMA (AutoRegressive Integrated Moving Average) model is specified using three parameters: p (order of the AR term), d (degree of differencing), and q (order of the MA term). We can use auto.arima to automatically select these parameters based on AIC.

auto.arima in R is designed to handle non-stationary data. It automatically applies the necessary differencing to make the series stationary before fitting the ARIMA model.

```r
arima_model <- auto.arima(train, stepwise = FALSE, trace = TRUE)

summary(arima_model)
```

```r
arima_forecast <- forecast(arima_model, h = length(test))
accuracy(arima_forecast, test)
```

The selected ARIMA(0, 1, 1) model uses one differencing step to handle non-stationarity and includes one moving average term. The training set metrics are favorable, but the model struggles on the test set, likely due to a significant change in the data pattern or a trend shift not captured by this simple model.

```r
# Figure : Plot the Auto ARIMA model with train and test data
autoplot(train, size = 1.25) +

  autolayer(arima_forecast, series = "Auto ARIMA", size = 1.25) +
  autolayer(test, series = "Test", size = 1.25) +

  ggtitle("AUTO ARIMA Model") +
  xlab("Time") +
  ylab("Price") +

  coord_cartesian(xlim = c(2020, 2024.5))+
  theme(
    legend.position = c(0.25, 0.75),
    legend.background = element_rect(fill = "white", color = NA),
    legend.title = element_text(size = 10),
    legend.text = element_text(size = 12) 
  )
```

## Custom ARIMA

Improving a custom ARIMA model often requires experimentation. By using a grid search on combinations of parameters, identify the optimal parameters for this specific time series data, that Selects parameters based on RMSE.

```r
# Initializing variables
best_rmse <- Inf
custom_arima_model <- NULL

# Creating the Grid Search/hyperparameter tuning of combinations
for (p in 0:2) {
    for (d in 0:2) {
        for (q in 0:2) {
            for (P in 0:1) {
                for (D in 0:1) {
                    for (Q in 0:1) {
                        model <- Arima(train, order = c(p, d, q), seasonal = list(order = c(P, D, Q), period = 12))
                        forecasted <- forecast(model, h = length(test))
                        model_rmse <- accuracy(forecasted, test)["Test set", "RMSE"]
                        if (model_rmse < best_rmse) {
                            best_rmse <- model_rmse
                            custom_arima_model <- model
                        }
                    }
                }
            }
        }
    }
}

# Summary of custom ARIMA model
summary(custom_arima_model)
```

```r
# Evaluaing the custom ARIMA model
custom_arima_forecast <- forecast(custom_arima_model, h = length(test))
accuracy(custom_arima_forecast, test)
```

```r
# Figure : Plot the Custom ARIMA model with train and test data
autoplot(train, size = 1.25) +
  
  autolayer(custom_arima_forecast, series = "Custom ARIMA", size = 1.25) +
  autolayer(test, series = "Test", size = 1.25) +

  ggtitle("Custom ARIMA Model") +
  xlab("Time") +
  ylab("Price") +

  coord_cartesian(xlim = c(2020, 2024.5))+
  theme(
    legend.position = c(0.25, 0.75),
    legend.background = element_rect(fill = "white", color = NA),
    legend.title = element_text(size = 10),
    legend.text = element_text(size = 12) 
  )
```
